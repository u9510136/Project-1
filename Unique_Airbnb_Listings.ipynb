```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib

import warnings
warnings.filterwarnings('ignore')

chi_data_to_load = "Resources/chi_listings.csv"
aus_data_to_load = "Resources/aus_listings.csv"
bos_data_to_load = "Resources/bos_listings.csv"
sea_data_to_load = "Resources/sea_listings.csv"
sd_data_to_load = "Resources/sd_listings.csv"

chi_file = pd.read_csv(chi_data_to_load)
aus_file = pd.read_csv(aus_data_to_load)
bos_file = pd.read_csv(bos_data_to_load)
sea_file = pd.read_csv(sea_data_to_load)
sd_file = pd.read_csv(sd_data_to_load)

chi_file.head()
```

```python
chi_unique = len(chi_file["id"].unique())
aus_unique = len(aus_file["id"].unique())
bos_unique = len(bos_file["id"].unique())
sea_unique = len(sea_file["id"].unique())
sd_unique = len(sd_file["id"].unique())
chi_unique
```

```python
city_listing_count = [chi_unique, aus_unique, bos_unique, sea_unique, sd_unique]
city_listing_count
```

```python
x_axis = np.arange(len(city_listing_count))
plt.figure(dpi=300)
plt.bar(x_axis, city_listing_count, alpha=0.5, color='purple', align="center")
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, ["Chicago", "Austin", "Boston", "Seattle", "San Diego"])
plt.xlabel("City Name")
plt.ylabel("Unique Airbnb Listings")
plt.title("Unique Airbnb Listings Per City")
plt.savefig("images/unique_airbnb_listings_per_city.png", bbox_inches="tight")
plt.show()
```
